```python
#create a variable called age
#since the maximum value of date is 31st dec 2013 lets caluclate age by 2014
df['Age']= 2014 - df['Year_Birth']
df['Age'].head()
```

```python
# creating spending variable
df['Spending']= df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
df['Spending'].head()
```

```python
df['Education'].unique()
```

```python

for i in df['Education']:
    if i == '2n Cycle':
        df['Education'].replace('2n Cycle','Bachelors',inplace= True)
    elif i == 'Graduation':
        df['Education'].replace('Graduation','Master',inplace= True)
    elif i == 'Basic':
        df['Education'].replace('Basic','Highschool',inplace= True)
```

```python
df['Education'].unique()
```

```python
df['Marital_Status'].unique()
```

```python
for i in df['Marital_Status']:
    if i == 'Together':
        df['Marital_Status'].replace('Together','Married',inplace= True)
    elif i == 'Absurd':
        df['Marital_Status'].replace('Absurd','Single',inplace= True)
    elif i == 'YOLO':
        df['Marital_Status'].replace('YOLO','Single',inplace= True)
    elif i == 'Alone':
        df['Marital_Status'].replace('Alone','Single',inplace= True)
    elif i == 'Divorced':
        df['Marital_Status'].replace('Divorced','Single',inplace= True)
    elif i == 'Widow':
        df['Marital_Status'].replace('Widow','Single',inplace= True)
```

```python
df['Marital_Status'].unique()
```

```python
df['Marital_Status'].head()
```

```python
def people(status):
  if status == 'Single':
    return 1
  else:
    return 2

df['People'] = df['Marital_Status'].apply(people)
```

```python
df['People'].head()
```

```python
# calculating total number of people at home
df['Total_People']= df['People'] + df['Kidhome'] + df['Teenhome']
df['Total_People'].head()
```

```python
df.head()
```

```python
df['Dt_Customer']= pd.to_datetime(df['Dt_Customer'],format= '%d-%m-%Y',dayfirst= True)
df['Dt_Customer'].head()
```

```python
#creating a reference date
current_date= pd.to_datetime('07-01-2014')
current_date
```

```python
#creating no.of seniority feature
df['Dt_Customer']= pd.to_datetime(df['Dt_Customer'],format= '%d-%m-%Y')
df['Seniority'] = (current_date - df['Dt_Customer']).dt.days
df['Seniority'].head()
```

```python
# Label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Education'] = le.fit_transform(df['Education'])
df['Marital_Status']= le.fit_transform(df['Marital_Status'])
```

```python
df.columns
```

```python
# caculating inter quartile range
q1= df['Income'].quantile(0.25)
q3= df['Income'].quantile(0.75)
iqr= q3-q1
print('Inter quartile range:' , iqr,'lower bound: ' , q1, 'upper bound:' , q3)
```

```python
# remove outliers using iqr and income columns

lower_lim = q1 - 1.5 * iqr
upper_lim = q3 + 1.5 * iqr
df = df[(df['Income'] > lower_lim) & (df['Income'] < upper_lim)]
```

```python
df.reset_index(drop= True, inplace= True)
```

```python
data= df[['Age','Education','Marital_Status','Income','Total_People','Seniority','Spending','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']]
data.head()
```

```python
data.columns
```

```python
# correlation matrix heatmat

import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
data['Income']= ss.fit_transform(data[['Income']])
data['Seniority']= ss.fit_transform(data[['Seniority']])
data['Spending']= ss.fit_transform(data[['Spending']])
```

```python
# elbow curve to get k value
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Apply K-means (assuming k=3 from elbow method)
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(data)

# Add cluster labels to the dataframe
data['Cluster'] = clusters

# Analyze results
print(data.groupby('Cluster').mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 2D scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(data=data, x='Age', y='Income', hue='Cluster', palette='deep')
plt.title('Clusters by Age and Income')
plt.show()

# 3D scatter plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(data['Age'], data['Income'], data['Spending'], c=data['Cluster'], cmap='viridis')
ax.set_xlabel('Age')
ax.set_ylabel('Income')
ax.set_zlabel('Spending')
plt.colorbar(scatter)
plt.title('3D Cluster Visualization')
plt.show()
```

```python
# Apply Agglomerative Clustering
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
agg_clustering = AgglomerativeClustering(n_clusters=3)  # You can adjust the number of clusters
clusters = agg_clustering.fit_predict(data)

# Add cluster labels to the dataframe
data['Cluster'] = clusters

# Visualize dendrogram
plt.figure(figsize=(10, 7))
dendrogram = shc.dendrogram(shc.linkage(data, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Euclidean Distance')
plt.show()

# Analyze results
print(data.groupby('Cluster').mean())

# Visualize clusters (example with 2D scatter plot)
plt.figure(figsize=(10, 8))
plt.scatter(data['Age'], data['Income'], c=data['Cluster'], cmap='viridis')
plt.title('Clusters by Age and Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.colorbar()
plt.show()
```

```python
#using cohen's d differencing to interprt the clusters
import numpy as np
import pandas as pd



numerical_columns = ['Age', 'Income', 'Total_People', 'Seniority', 'Spending',
                     'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
                     'MntSweetProducts', 'MntGoldProds']

def cohens_d(cluster1, cluster2):
    n1, n2 = len(cluster1), len(cluster2)
    var1, var2 = np.var(cluster1, ddof=1), np.var(cluster2, ddof=1)
    s = np.sqrt(((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2))
    d = (np.mean(cluster1) - np.mean(cluster2)) / s
    return d
for feature in numerical_columns:
    print(f"\nCohen's d for {feature}:")
    for i in range(3):
        for j in range(i+1, 3):
            d = cohens_d(data[data['Cluster'] == i][feature],
                         data[data['Cluster'] == j][feature])
            print(f"  Between clusters {i} and {j}: {d:.2f}")
def interpret_clusters(data, numerical_columns):
    for i in range(3):
        print(f"\nCluster {i} characteristics:")
        for feature in numerical_columns:
            d_values = [cohens_d(data[data['Cluster'] == i][feature],
                                 data[data['Cluster'] == j][feature])
                        for j in range(3) if j != i]
            avg_d = np.mean(np.abs(d_values))

            if avg_d >= 0.8:
                direction = "high" if np.mean(d_values) > 0 else "low"
                print(f"- {direction.capitalize()} {feature} (avg d = {avg_d:.2f})")

interpret_clusters(data, numerical_columns)
```
